## Feature Scaling
---

**Algunos modelos comúnmente utilizados**
  
|Nombre del algoritmo | Necesita escalamiento? |
|--|--|
| SVM, KNN, PCA, K-Means, Linear Regression, Logistic Regression, Linear Discriminant Analysis| $\checkmark$ |
| Naive Bayes, Tree-Based models | $\times$ |

* Asimetría afecta modelos PCA, entonces es recomendable utilizar Power Transformations para remover asimetría.

### Min-Max scaling

Supongamos que queremos re-escalar el atributo en el rango $(a,b)$  

Fórmula: ${x}' = a + \frac{x - \mathrm{min}(x)\times (b - a) }{\mathrm{max}(x) -\mathrm{min}(x)}$  

Éste método es muy sensitivo a los valores atípicos porque éstos afectan tanto min$(x)$ como max$(x)$

```python
from sklearn.preprocessing import MinMaxScaler  
```

```python
mm_scaler = MinMaxScaler()
X_train['AveRooms_mm'] = mm_scaler.fit_transform(X_train[['AveRooms']])
X_train_99['AveRooms_mm'] = mm_scaler.fit_transform(X_train_99[['AveRooms']])
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(16,6))
sns.histplot(data=X_train['AveRooms'],ax=axes[0], kde=True)
sns.histplot(data=X_train_99['AveRooms_mm'],ax=axes[1], kde=True)
```

### Standardization or Normalization

Fórmula: ${z} = \frac{x - \mu }{\sigma}$ 

Donde $\mu$ es la media de $X$ y $\sigma$ es la desviación estándard.

Éste método también es sensitivo a valores atípicos porque afectan tanto $\mu$ como $\sigma$

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss_scaler = StandardScaler()
X_train['AveRooms_ss'] = ss_scaler.fit_transform(X_train[['AveRooms']])
X_train_99['AveRooms_ss'] = ss_scaler.fit_transform(X_train_99[['AveRooms']])
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(16,6))
sns.histplot(data=X_train['AveRooms'],ax=axes[0], kde=True)
sns.histplot(data=X_train_99['AveRooms_ss'],ax=axes[1], kde=True)
```

### $l^2$ Normalization

Éste método trabaja sobre las filas del vector de atributos, no sobre columnas separadas. 

Supongamos que X[i][j] representa el valor del atributo $j$ en la instancia $i$.  
Fórmula para la norma $l^1$: $X[i][j] = \frac{X[i][j]}{\sum_{j=1|X[i][j]|}}$  
Fórmula para la norma $l^2$: $X[i][j] = \frac{X[i][j]}{\sqrt{\sum_{j=1}X[i][j]^2}}$  


```python
from sklearn.preprocessing import Normalizer  
```

```python
l1_scaler = Normalizer(norm='l1')
l2_scaler = Normalizer()
```

```python
xtrain_norm = X_train_99[['AveRooms','MedInc']]
```

```python
xtrain_norm1 = l1_scaler.fit_transform(xtrain_norm)
xtrain_norm2 = l2_scaler.fit_transform(xtrain_norm)
```

```python
fig, axes = plt.subplots(ncols=3,figsize=(21,6))
sns.scatterplot(data=X_train_norm, x='AveRooms',y='MedInc',ax=axes[0])
sns.scatterplot(x=X_train_norm1[:,0],y=X_train_norm1[:,1],ax=axes[1])
sns.scatterplot(x=X_train_norm2[:,0],y=X_train_norm2[:,1],ax=axes[2])
```

### Robust Scaling

Supongamos que queremos escalar el atributo en el rango de cuantiles $(a, b)$

Fórmula: ${x}' = \frac{x - \mathrm{median}(x)}{\mathrm{quantile}(x,b) - \mathrm{quantile}(x,a)}$  

Este método es más robusto a valores atípicos.

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs_scaler = RobustScaler()
X_train['AveRooms_rs'] = rs_scaler.fit_transform(X_train[['AveRooms']])
X_train_99['AveRooms_rs'] = rs_scaler.fit_transform(X_train_99[['AveRooms']])
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(16,6))
sns.histplot(data=X_train['AveRooms'],ax=axes[0], kde=True)
sns.histplot(data=X_train_99['AveRooms'],ax=axes[1], kde=True)
```

## Feature Transformation
---

### Transformación Box-Cox

La transformación Box-Cox funciona únicamente para números estrictamente positivos. La transformación esta definida de la siguiente manera:

$x_i'(\lambda) = \left\{\begin{matrix}
 \frac{x_i^\lambda - 1}{\lambda}& if \lambda \neq 0\\ 
 log(\lambda)& if \lambda =  0
\end{matrix}\right.$  

### Transformación Yeo-Johnson

La transformación Yeo-Johnson trabaja tanto para valores positivos como negativos. La transformación esta definida de la siguiente manera:

$x_i(\lambda) = \left\{\begin{matrix}
\frac{[(x_i + 1)^{\lambda} -1]}{\lambda} & if \, \, \lambda\neq 0, x_i \geq  0 \\ 
ln(x_i + 1) & if \, \, \lambda =  0, x_i \geq  0\\ 
-\frac{[(-x_i + 1)^{2- \lambda} -1]}{2-\lambda} & if \, \, \lambda\neq 2, x_i <  0 \\ 
-ln(-x_i + 1) & if \, \, \lambda =  2, x_i <  0\\ 
\end{matrix}\right.$  


Usualmente los valores de $\lambda$  entre $[-5,5]$ son considerados y el valor óptimo que estabiliza la varianza y minimiza la asimetría es seleccionado utilizando MLE (Maximum Likelihood Estimate)

```python
from sklearn.preprocessing import PowerTransformer
```

```python
bx_transformerc = PowerTransformer(method='box-cox')
bx_transformerj = PowerTransformer() #Default = Yeo-Johnson
```

```python
##Box-cox
X_train['MedInc_bx'] = bx_transformerc.fit_transform(X_train[['MedInc']])
X_train_99['MedInc_bx'] = bx_transformerc.fit_transform(X_train_99[['MedInc']])

##Yeo-Johnson
X_train['MedInc_yj'] = bx_transformerj.fit_transform(X_train[['MedInc']])
X_train_99['MedInc_yj'] = bx_transformerj.fit_transform(X_train_99[['MedInc']])
```

```python
##Box-cox
fig, axes = plt.subplots(ncols=2, figsize=(16,6))
sns.histplot(data=X_train['MedInc'],ax=axes[0], kde=True)
sns.histplot(data=X_train_99['MedInc_bx'],ax=axes[1], kde=True)
```

## Feature Discretization
---

### Binarization

Convertir variables numéricas en respuestas binarias

```python
from sklearn.preprocessing import Binarizer
```

```python
binarizer = Binarizer(threshold=6)
```

```python
X_train['MedInc_bin'] = binarizer.fit_transform(X_train[['MedInc']])
X_train_99['MedInc_bin'] = binarizer.fit_transform(X_train_99[['MedInc']])
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(16,6))
sns.histplot(data=X_train['MedInc'],ax=axes[0], kde=True)
sns.countplot(x=X_train_99['MedInc_bin'],ax=axes[1])
```

### Uniform binning

Mapear variables numéricas en bloques uniformes

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
nbins = 5
ubin_discretizer = KBinsDiscretizer(n_bins= nbins, encode='ordinal', strategy='uniform')
```

```python
X_train['Population_ubin'] = ubin_discretizer.fit_transform(X_train[['Population']])
X_train_99['Population_ubin'] = ubin_discretizer.fit_transform(X_train_99[['Population']])
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(6,8))
sns.histplot(data=X_train['Population'],ax=axes[0], kde=True)
for edge in ubin_discretizer.bin_edges_[0]:
    line = axes[0].axvline(edge,color='b')
axes[0].legend([line],['Uniform in edges'], fontsize = 10)
sns.countplot(x=X_train_99['Population_ubin'])
axes[1].set_title(f'Uniform binning resul ((n_bins) bins)')
plt.tight_layout()
```

### *k*-means binning


Mapear variables numéricas en *k* bloques utilizando técnicas de agrupamiento o clustering

```python
nbins = 5
kbin_discretizer = KBinsDiscretizer(n_bins= nbins, encode='ordinal', strategy='kmeans')
```

```python
X_train['Population_kbin'] = kbin_discretizer.fit_transform(X_train[['Population']])
X_train_99['Population_kbin'] = kbin_discretizer.fit_transform(X_train_99[['Population']])
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(6,8))
sns.histplot(data=X_train['Population'],ax=axes[0], kde=True)
axes[0].set_title('Histogram with whole training data')
for edge in kbin_discretizer.bin_edges_[0]:
    line = axes[0].axvline(edge,color='b')
axes[0].legend([line],['K-Means bin edges'], fontsize = 10)
sns.countplot(x=X_train_99['Population_ubin'])
axes[1].set_title(f'K-Means binning resul ((n_bins) bins)')
plt.tight_layout()
```
